```python
import pandas
```

```python
Revenue_Growth_Factor = 1.18
EBITDA_Margin = 0.45
Taxes = 0.21
Capital_Expenditures = 0.03
Working_Capital_Increase = 0.06
EBITDA_Exit_Multiple = 32.70
Discount_Factor = 0.94
```

```python
dates = pandas.date_range(start='2022', end='2031', freq='Y')
dates
```

```python
df = pandas.DataFrame(index=dates)
terminal_vaule_df = pandas.DataFrame()
df
```

```python
revenue =222

rev = [revenue * Revenue_Growth_Factor]

for i in range(1,len(dates)):
    rev.append(rev[i-1]*Revenue_Growth_Factor)
        
rev
```

```python
df['Revenue'] = rev
df
```

```python
df['EBITDA'] = df['Revenue']*EBITDA_Margin
df
```

```python
df["Taxes"] = df ["EBITDA"] * -Taxes
df
```

```python
df['Capital Expenditure'] = df['Revenue'] * -Capital_Expenditures
df
```

```python
df["Working Capital"] = df["Revenue"] * -Working_Capital_Increase
df
```

```python
df['cash_flow'] = df.drop(columns=['Revenue']).sum(axis=1)
df
```

```python
df['Discount_Factor']=[Discount_Factor ** i for i in range(1,len(dates)+1)] 
df
```

```python
df.transpose()
```

```python
discount_factors= [Discount_Factor ** i for i in range(1,len(dates)+1)]
Discount_table = pandas.DataFrame(index=discount_factors)
Discount_table['present_value_Future_Cash_Flow'] = discount_factors * df['cash_flow'].values
Discount_table
```

```python
present_value = Discount_table['present_value_Future_Cash_Flow'].sum()
```

```python
present_value.round()
```

```python
terminal_frame = pandas.DataFrame(index=['2032'])
terminal_frame['Revenue'] = df['Revenue'].iloc[-1] * Revenue_Growth_Factor
terminal_frame['EBITDA'] =terminal_frame['Revenue'] * EBITDA_Margin
```

```python
rv= terminal_frame['Revenue'].round()
eb = terminal_frame[['EBITDA']].round()
```

```python
t_v= terminal_frame['terminal_value'] = terminal_frame['EBITDA'] * EBITDA_Exit_Multiple + present_value
terminal_frame['terminal_value'].round()
```

```python
print(f"""
Our projected Present Value of Projected Future Cash Flows are: ${present_value.round()}.
We are providing revenue guidance in 2030 to be: ${rv.values}.
With an estimated EBITDA around: ${eb.values}.

The terminal value a this time given a multiple of {EBITDA_Exit_Multiple} and the above EBITDA for 2030, would equal:{t_v.values.round()}
""")
```
